# <span style="color:Blue">First order monotone schemes  </span>

## <span style="color:red">One-Dimentional case </span>

We begin with N = 1 and write HJ in the form : 

$$ \partial _{t}u + H\left ( \partial _{x}u \right ) =0 \quad u(x,0)=u_{0}(x) \quad \text{for } t>0 \text{ , } x \in \left [ a , b \right ]
\quad (1) $$  

#### Notations

Given mesh sizes $\Delta x \text{ , } \Delta t > 0$, our numerical approximation at $(x_{j}, t^{n})$ will be denoted $ U_{j}^{n} \quad \forall j,n \in \mathbb{Z} $. Where $x_{j}= j \Delta x \text{ , } t^{n}= n \Delta t$


$ \lambda^{x} = \frac{\Delta t}{\Delta x} \text{ , } \Delta _{+}^{x} U_{j} = U_{j+1} - U_{j} \text{ , } p^{+}_{j} =\frac{ \Delta _{+}^{x} U_{j} }{\Delta x} \text{ , } p^{-}_{j} =\frac{\Delta _{+}^{x} U_{j-1}}{\Delta x}$  

#### Note

Since we are in 1D and $\Omega =\mathbb{R}$, there is a relation with conservation laws.

#### Scheme

Our finite difference schemes are in the form :

$$ U_{j}^{n+1} = U_{j}^{n} - \Delta t H^{num} \left ( p^{-} , p^{+}  \right )$$

We define our $S_{h}$ operator : $h=(\Delta t , \Delta x)$ 
$$S_{h}(x_{j},t^{n+1},U_{j}^{n+1},u^{h})=\frac{U_{j}^{n+1}-U_{j}^{n}}{\Delta t} + H^{num}(x_{j},\frac{U_{j}^{n}-U_{j-1}^{n}}{\Delta x},\frac{U_{j+1}^{n}-U_{j}^{n}}{\Delta x})$$

#### Properties of $H^{num}$

- Lipschitz continuous.
- Consistent with $H$ : $H^{num} (p,p)= H(p)$
- Monotone : $H^{num}\left ( \uparrow , \downarrow  \right )$  (the monotonicity means that the map ($U^{n} \rightarrow U_{j}^{n+1}$) is a nondecreasing function of each $U_{i}^{n}$).
- Stability ?? 

We now give some examples of monotone fluxes $H^{num}$

###  Local Lax-Friedrichs Scheme (LLF)

The numerical Hamiltonian is given by 
$$ H^{LLF} (p^{-} , p^{+}) = H(\frac{p^{-} +p^{+}}{2})-\frac{1}{2}\alpha ^{x}(p^{+} - p^{-} )  \quad \text{for } p^{-} , p^{+} \in \mathbb{R} $$
Where 
$$\alpha ^{x}= \underset{p^{*} \in I(p^{-},p^{+})}{max}\left | \partial_{p} H(p^{*}) \right |$$
And 
$$
I(p^{-},p^{+})=\left [ \min\left ( p^{-},p^{+} \right ),  \max\left ( p^{-},p^{+} \right ) \right ]$$

### Godunov Scheme

The numerical Hamiltonian :
    $$ H^{G}(p^{-},p^{+})= \left\{\begin{matrix}
\underset{p^{} \in I(p^{-},p^{+})}{min} \left \{ H(p) \right \} & \text{ if } p^{-}\leq p^{+} \\ 
\underset{p^{} \in I(p^{-},p^{+})}{max} \left \{ H(p) \right \} & \text{ if } p^{-}\geq p^{+} 
\end{matrix}\right.$$

### Roe Scheme with LLF entropy correction

The numerical Hamiltonian :
$$ H^{RF}(p^{-},p^{+}) = \left\{\begin{matrix}
H(p^{*}) & \text{ if } \partial _{p}H(p) \text{ doesn't change sign in } I(p^{-},p^{+}) \\ 
H^{LF}(p^{-},p^{+}) & \text{ else }
\end{matrix}\right.$$
Where
$$p^{*}=\left\{\begin{matrix}
p^{-} & \text{ if } \partial _{p}H(p) \geq 0 \\ 
p^{+} & \text{ else }
\end{matrix}\right.$$

## <span style="color:red">Two-Dimentional case </span>

We take N = 2 and write HJ in the form : 

$$ \partial _{t}u + H\left ( \partial _{x}u , \partial _{y}u \right ) =0 \quad u(x,y,0)=u_{0}(x,y) \quad \text{for } t>0 \text{ , } x,y \in \left [ a , b \right ]
\quad (1) $$  

#### Notations

Given mesh sizes $\Delta x \text{ , } \Delta y \text{ , } \Delta t > 0$, our numerical approximation at $(x_{i}, y_{j}, t^{n})$ will be denoted $ U_{i,j}^{n} \quad \forall i,j,n \in \mathbb{Z} $. Where $x_{i}= i \Delta x \text{ , } y_{j}= j \Delta y \text{ , } t^{n}= n \Delta t$


$ \lambda^{x} = \frac{\Delta t}{\Delta x} \text{ , } \lambda^{y} = \frac{\Delta t}{\Delta y} \text{ , } \Delta _{+}^{x} U_{i,j} = U_{i+1,j} - U_{i,j} \text{ , } \Delta _{+}^{y} U_{i,j} = U_{i,j+1} - U_{i,j}$  
$p^{+}_{i} =\frac{ \Delta _{+}^{x} U_{i,j} }{\Delta x} \text{ , } p^{-}_{i} =\frac{\Delta _{+}^{x} U_{i-1,j}}{\Delta x} q^{+}_{j} =\frac{ \Delta _{+}^{y} U_{i,j} }{\Delta y} \text{ , } q^{-}_{j} =\frac{\Delta _{+}^{y} U_{i,j-1}}{\Delta y}$  

#### Scheme

Our finite difference schemes are in the form :

$$ U_{i,j}^{n+1} = U_{i,j}^{n} - \Delta t H^{num} \left ( p^{-}_{i} , p^{+}_{i} , q^{-}_{j} , q^{+}_{j}  \right )$$

We define our $S_{h}$ operator : $h=(\Delta t , \Delta x , \Delta y)$ 
$$S_{h}(x_{i},y_{i},t^{n+1},U_{i,j}^{n+1},u^{h}) =\frac{U_{i,j}^{n+1}-U_{i,j}^{n}}{\Delta t} + H^{num}(x_{i},y_{j},\frac{U_{i,j}^{n}-U_{i-1,j}^{n}}{\Delta x},\frac{U_{i+1,j}^{n}-U_{i,j}^{n}}{\Delta x},\frac{U_{i,j}^{n}-U_{i,j-1}^{n}}{\Delta y},\frac{U_{i,j+1}^{n}-U_{i,j}^{n}}{\Delta y})$$

#### Properties of $H^{num}$

- Lipschitz continuous.
- Consistent with $H$ : $H^{num} (p,p,q,q)= H(p,q)$
- Monotone : $H^{num}\left ( \uparrow , \downarrow , \uparrow , \downarrow  \right )$  (the monotonicity means that the map ($U^{n} \rightarrow U_{i,j}^{n+1}$) is a nondecreasing function of each $U_{i,j}^{n}$).
- Stability ?? 

We now give some examples of monotone fluxes $H^{num}$

###  Local Lax-Friedrichs Scheme (LLF)

The numerical Hamiltonian is given by 
$$ H^{LLF} (p^{-} , p^{+} , q^{-} , q^{+}) = H(\frac{p^{-} +p^{+}}{2} , \frac{q^{-} +q^{+}}{2})-\frac{1}{2}\alpha ^{x}(p^{+} - p^{-} )-\frac{1}{2}\alpha ^{y}(q^{+} - q^{-} )  \quad \text{for } p^{-} , p^{+} , q^{-} , q^{+} \in \mathbb{R} $$
Where 
$$\alpha ^{x}= \underset{p^{*} \in I(p^{-},p^{+}) \\ \forall q^{*}}{max}\left | \partial_{p} H(p^{*},q^{*}) \right | \text{ and } \alpha ^{y}= \underset{\forall p^{*} \\ q^{*} \in I(q^{-},q^{+})}{max}\left | \partial_{q} H(p^{*},q^{*}) \right |$$
And 
$$
I(r^{-},r^{+})=\left [ \min\left ( r^{-},r^{+} \right ),  \max\left ( r^{-},r^{+} \right ) \right ]$$

### Godunov Scheme

The numerical Hamiltonian :
$$H^{G}(p^{-},p^{+},q^{-},q^{+})= \underset{p \in I(p^{-},p^{+})}{ext} \: \: \underset{q \in I(q^{-},q^{+})}{ext} H(p,q)$$

With 
    $$ \underset{r \in I(r^{-},r^{+})}{ext} = \left\{\begin{matrix}
\underset{r \in I(r^{-},r^{+})}{min} & \text{ if } r^{-}\leq r^{+} \\ 
\underset{r \in I(r^{-},r^{+})}{max}  & \text{ if } r^{-}\geq r^{+} 
\end{matrix}\right.$$

### Roe Scheme with LLF entropy correction

The numerical Hamiltonian :
$$ H^{RF}(p^{-},p^{+},q^{-},q^{+}) = \left\{\begin{matrix}
H(p^{*},q^{*}) & \text{ if } \partial _{p}H(p,q), \partial _{q}H(p,q) \text{ do not change signs in } p\in I(p^{-},p^{+}), q\in I(q^{-},q^{+}) \\ 
H^{LLF}(p^{-},p^{+},q^{*},q^{*}) & \text { Otherwise and if } \partial _{q}H(p,q) \text{ does not change sign in } \forall p, q\in I(q^{-},q^{+}) \\
H^{LLF}(p^{*},p^{*},q^{-},q^{+}) & \text { Otherwise and if } \partial _{p}H(p,q) \text{ does not change sign in } p\in I(p^{-},p^{+}), \forall q \\
H^{LLF}(p^{-},p^{+},q^{-},q^{+}) & \text{ Otherwise }
\end{matrix}\right.$$
Where
$$p^{*}=\left\{\begin{matrix}
p^{-} & \text{ if } \partial _{p}H(p,q) \geq 0 \\ 
p^{+} & \text{ else }
\end{matrix}\right. \text{ and } q^{*}=\left\{\begin{matrix}
q^{-} & \text{ if } \partial _{q}H(p,q) \geq 0 \\ 
q^{+} & \text{ else }
\end{matrix}\right.$$
